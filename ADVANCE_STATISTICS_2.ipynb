 2.What is the null hypothesis, and how does it differ from the alternative hypothesis?

 ANS.The null hypothesis (H0) is a statement in scientific research that asserts there is no effect or no relationship between two variables in the population.
 It is often described as the hypothesis of no difference or no effect, and it is the hypothesis that researchers aim to either reject or fail to reject based on the evidence from the sample data.

In contrast, the alternative hypothesis (H1 or Ha) is the statement that there is an effect or a relationship between the variables.
 It is the hypothesis that researchers hope to support with their data, and it is the complement to the null hypothesis.
 The alternative hypothesis suggests that the observed effect in the sample data is not due to chance alone but rather indicates a real effect in the population.

The null hypothesis and the alternative hypothesis are mutually exclusive and exhaustive, meaning that only one can be true, and together they cover all possible outcomes.
 The null hypothesis is typically the status quo or the default assumption, while the alternative hypothesis represents the research hypothesis that the researcher is trying to prove.

Choosing the right significance level is essential for making accurate data-driven decisions, balancing the risk of false positives with the need to detect real effects.
 This level helps researchers decide whether to accept or reject the null hypothesis based on the p-value obtained from the statistical test.

The significance level is important because it helps control the error rate in statistical tests, ensuring that conclusions drawn from the data are reliable and valid.

  ANS.In hypothesis testing, a P-value represents the probability of observing your data, or something more extreme, if the null hypothesis is true.

 5.How do you interpret the P-value in hypothesis testing?

 ANS.In hypothesis testing, the p-value is the probability of obtaining test results at least as extreme as the result actually observed, under the assumption that the null hypothesis is correct.
 A low p-value suggests that the observed data are inconsistent with the null hypothesis, potentially favoring an alternative hypothesis.
 To interpret the p-value, you need to compare it to a chosen significance level (α), generally set at 5% (α = 0.05).
 If the p-value is less than or equal to the significance level (p ≤ α = 0.05), you reject the null hypothesis, indicating there is sufficient evidence to conclude that the observed effect or relationship is statistically significant and unlikely to have occurred by chance alone.
 Conversely, if the p-value is greater than the significance level (p > α = 0.05), you fail to reject the null hypothesis, meaning the observed effect or relationship does not provide enough evidence to reject the null hypothesis.
 This does not necessarily mean there is no effect; it simply means the sample data does not provide strong enough evidence to rule out the possibility that the effect is due to chance.

 6.What are Type 1 and Type 2 errors in hypothesis testing?

 ANS.In hypothesis testing, Type 1 and Type 2 errors refer to incorrect conclusions drawn about a population based on sample data. A Type 1 error, also known as a false positive, occurs when the null hypothesis is rejected even though it is true. This means that you conclude there is a notable effect or difference when there isn't one.

A Type 2 error, or false negative, happens when you fail to reject the null hypothesis when the alternative hypothesis is actually true. In other words, you miss detecting an effect or difference that does exist.

10.What is the T-distribution, and when should it be used instead of the normal distribution?

ANS.The t-distribution, also known as Student's t-distribution, is a probability distribution used to estimate population parameters when the sample size is small and the population standard deviation is unknown.

 This makes the t-distribution more conservative and appropriate for smaller sample sizes or when variances are unknown.

The t-distribution should be used instead of the normal distribution under the following conditions:

When the sample size is small (typically less than 30).
When the population standard deviation is unknown and must be estimated from the sample data.
When the data are approximately normally distributed but the population variance is unknown.
As the sample size increases, the t-distribution approaches the normal distribution.
 For large sample sizes (generally greater than 30), the normal distribution can be used instead.

 20.What is an ANOVA test, and what are its assumptions?

 ANS.An ANOVA (Analysis of Variance) test is a statistical method used to compare the means of three or more groups to determine if there are significant differences between them. It analyzes the variance within the groups and between groups to assess whether the observed variance between group means is more significant than within the groups. If the observed variance between group means is significant, it suggests that the differences are meaningful.

ANOVA relies on several key assumptions that must be met for the test results to be valid:

Independence of observations: The data were collected using statistically valid sampling methods, and there are no hidden relationships among observations.

Normality: The values of the dependent variable follow a normal distribution within each group.

Homogeneity of variance: The variation within each group being compared is similar for every group. This is also known as homoscedasticity.

21.What are the different types of ANOVA tests?

ANS.ANOVA, or Analysis of Variance, is a statistical method used to compare the means of three or more groups to determine if there are any significant differences between them. There are several types of ANOVA tests, each suited for different scenarios:

One-way ANOVA: This test is used when comparing the means of three or more independent groups based on one factor. It assesses differences in one continuous variable across one grouping variable. For example, it could be used to assess differences in job satisfaction levels between ethnicities.

Two-way ANOVA: Also known as factorial ANOVA, this test measures the interaction effect between two different groups (or independent variables) on a control group. It uses F-ratios with two mean square values related to each group.

Repeated Measures ANOVA: This ANOVA test takes sample means from at least three different sets of test statistics and compares them against one another. It is used when the same participants are measured multiple times under different conditions or time points.

MANOVA (Multivariate Analysis of Variance): This form of test is ideal if you plan to take a high number of categorical variables into account. The acronym MANOVA stands for multivariate analysis of variance. It can handle up to seven independent variables, leading to significant differences worth keeping track of in each case.

Two-way ANOVA with Replication: This test studies the effect size of two separate datasets, requiring you to run multiple studies with different groups of people, while still using the same response variables.

2.Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define the range for the discrete uniform distribution
low = 1   # inclusive
high = 7  # exclusive (like a 6-sided die: 1 to 6)

# Create the distribution
dist = randint(low, high)

# Define the values (support) and their corresponding PMF
x = np.arange(low, high)
pmf_values = dist.pmf(x)

# Plot the PMF
# Removed the deprecated 'use_line_collection' argument
plt.stem(x, pmf_values)
plt.title('PMF of Discrete Uniform Distribution (1 to 6)')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.grid(True)
plt.show()
```

3.Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

```python
def bernoulli_pmf(x, p):
    """
    Calculate the probability mass function of a Bernoulli distribution.

    Parameters:
    - x (int): Outcome (must be 0 or 1)
    - p (float): Probability of success (0 ≤ p ≤ 1)

    Returns:
    - float: PMF value at x
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1 for a Bernoulli distribution.")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1.")

    return p if x == 1 else 1 - p

# Example usage
print("P(X=0):", bernoulli_pmf(0, 0.3))
print("P(X=1):", bernoulli_pmf(1, 0.3))
```

6.Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
uniform distribution>.

```python
import numpy as np
import matplotlib.pyplot as plt

def discrete_uniform_cdf(a, b):
    """
    Computes the values and CDF of a discrete uniform distribution over integers a to b.

    Args:
        a (int): Lower bound (inclusive)
        b (int): Upper bound (inclusive)

    Returns:
        x (ndarray): Discrete values
        cdf (ndarray): Cumulative probabilities
    """
    x = np.arange(a, b + 1)
    n = b - a + 1
    cdf = np.arange(1, n + 1) / n
    return x, cdf

# Parameters for the discrete uniform distribution
a = 1  # Lower bound
b = 10 # Upper bound
# Calculate CDF
x, cdf = discrete_uniform_cdf(a, b)

# Plot CDF
plt.step(x, cdf, where='post', label='CDF', color='blue')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.title('CDF of Discrete Uniform Distribution')
plt.xticks(x)
plt.yticks(np.linspace(0, 1, b - a + 2))
plt.grid(True)
plt.legend()
plt.show()
```

9.Write a Python function to calculate Z-scores from a dataset and plot them.

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """
    Calculates Z-scores for a given dataset.

    Args:
        data (array-like): Input numeric data.

    Returns:
        z_scores (ndarray): Z-scores of the data.
    """
    mean = np.mean(data)
    std = np.std(data)
    z_scores = (data - mean) / std
    return z_scores

def plot_z_scores(z_scores):
    """
    Plots a histogram of Z-scores.
Args:
        z_scores (array-like): Z-scores to be plotted.
    """
    plt.hist(z_scores, bins=50, density=True, alpha=0.7, color='teal', edgecolor='black')
    plt.title('Histogram of Z-Scores')
    plt.xlabel('Z-Score')
    plt.ylabel('Density')
    plt.grid(True)
    plt.axvline(0, color='red', linestyle='--', label='Mean (Z=0)')
    plt.legend()
    plt.show()

# Example usage
data = np.random.normal(50, 10, 1000)  # Simulated dataset
z_scores = calculate_z_scores(data)
plot_z_scores(z_scores)
```

10.Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_clt(distribution_func, dist_params, sample_size=30, num_samples=1000):
    """
    Simulate the Central Limit Theorem using a non-normal distribution.

    Args:
        distribution_func (callable): NumPy distribution function (e.g., np.random.exponential).
        dist_params (dict): Parameters for the distribution function.
        sample_size (int): Size of each sample.
        num_samples (int): Number of sample means to compute.

    Returns:
        sample_means (ndarray): Array of sample means.
    """
    sample_means = []
    for _ in range(num_samples):
        sample = distribution_func(size=sample_size, **dist_params)
        sample_means.append(np.mean(sample))
    return np.array(sample_means)
# Parameters
distribution = np.random.exponential
params = {"scale": 2.0}  # mean = scale for exponential
sample_size = 30
num_samples = 1000

# Simulate
means = simulate_clt(distribution, params, sample_size, num_samples)

# Plot
plt.hist(means, bins=40, density=True, alpha=0.7, color='steelblue', edgecolor='black')
plt.title(f'CLT Demonstration (Sample Size = {sample_size})')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.grid(True)
plt.axvline(np.mean(means), color='red', linestyle='--', label='Mean of Sample Means')
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy.stats import norm

def z_test(sample, population_mean, population_std, alpha=0.05, alternative='two-sided'):
    """
    Perform Z-test for hypothesis testing.

    Parameters:
    - sample: list or numpy array of sample data
    - population_mean: hypothesized population mean
    - population_std: population standard deviation (assumed known)
    - alpha: significance level (default 0.05)
    - alternative: 'two-sided', 'less', or 'greater'
    """
    n = len(sample)
    sample_mean = np.mean(sample)
    standard_error = population_std / np.sqrt(n)
    z = (sample_mean - population_mean) / standard_error

    print(f"\nSample mean: {sample_mean:.4f}")
    print(f"Z-statistic: {z:.4f}")

    if alternative == 'two-sided':
        p_value = 2 * (1 - norm.cdf(abs(z)))
        reject = p_value < alpha
        print(f"P-value (two-tailed): {p_value:.4f}")
    elif alternative == 'greater':
        p_value = 1 - norm.cdf(z)
        reject = p_value < alpha
        print(f"P-value (right-tailed): {p_value:.4f}")
    elif alternative == 'less':
        p_value = norm.cdf(z)
        reject = p_value < alpha
        print(f"P-value (left-tailed): {p_value:.4f}")
    else:
        raise ValueError("alternative must be 'two-sided', 'less', or 'greater'")

    if reject:
        print("Result: Reject the null hypothesis.")
    else:
        print("Result: Fail to reject the null hypothesis.")

# Example usage
if __name__ == "__main__":
    # Sample dataset (e.g., test scores)
    sample_data = [72, 75, 78, 70, 69, 74, 73, 71, 77, 76]

    # Known population parameters
    pop_mean = 70     # Hypothesized population mean (H₀)
    pop_std = 5       # Known population standard deviation

    # Perform a two-tailed Z-test
    z_test(sample_data, pop_mean, pop_std, alpha=0.05, alternative='two-sided')
```

16.Create a confidence interval for a dataset using Python and interpret the result.

```python
import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
    - data: list or numpy array of data
    - confidence: confidence level (default is 95%)

    Returns:
    - (lower_bound, upper_bound)
    """
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # standard error of the mean
    df = n - 1  # degrees of freedom
    t_crit = stats.t.ppf((1 + confidence) / 2, df)

    margin_of_error = t_crit * std_err
    lower = mean - margin_of_error
    upper = mean + margin_of_error

    print(f"\nSample size: {n}")
    print(f"Sample mean: {mean:.4f}")
    print(f"Standard error: {std_err:.4f}")
    print(f"Confidence level: {confidence*100:.1f}%")
    print(f"Confidence interval: ({lower:.4f}, {upper:.4f})")

    return (lower, upper)

# Example usage
if __name__ == "__main__":
    # Sample data (e.g., weights, scores, measurements)
    data = [72, 75, 78, 70, 69, 74, 73, 71, 77, 76]
    confidence_interval(data, confidence=0.95)
```

```python
from scipy.stats import norm
import numpy as np

def z_test(sample, pop_mean, pop_std, alpha=0.05):
    """
    Perform one-sample Z-test.

    Parameters:
        sample (array-like): sample data
        pop_mean (float): population mean
        pop_std (float): population standard deviation (known)
        alpha (float): significance level (default 0.05)

    Returns:
        None
    """
    n = len(sample)
    sample_mean = np.mean(sample)

    # Calculate Z statistic
    z = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))

    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z)))

    print(f"Sample mean = {sample_mean:.4f}")
    print(f"Population mean = {pop_mean}")
    print(f"Z statistic = {z:.4f}")
    print(f"P-value = {p_value:.4f}")

    if p_value < alpha:
        print(f"Reject the null hypothesis at α = {alpha}")
        print("=> Sample mean is significantly different from population mean.")
    else:
        print(f"Fail to reject the null hypothesis at α = {alpha}")
        print("=> No significant difference between sample mean and population mean.")
```

```python
import numpy as np
from scipy.stats import norm

def two_sample_z_test(sample1, sample2, pop_std1, pop_std2, alpha=0.05):
    """
    Perform two-sample Z-test for difference of means with known population std dev.

    Parameters:
        sample1, sample2: array-like sample data
        pop_std1, pop_std2: known population std deviations
        alpha: significance level

    Returns:
        None
    """
    n1, n2 = len(sample1), len(sample2)
    mean1, mean2 = np.mean(sample1), np.mean(sample2)

    # Calculate pooled standard error
    se = np.sqrt((pop_std1**2)/n1 + (pop_std2**2)/n2)

    # Z statistic
    z = (mean1 - mean2) / se

    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z)))

    print(f"Sample 1 mean = {mean1:.4f}, n = {n1}")
    print(f"Sample 2 mean = {mean2:.4f}, n = {n2}")
    print(f"Z statistic = {z:.4f}")
    print(f"P-value = {p_value:.4f}")

    if p_value < alpha:
        print(f"Reject null hypothesis at α = {alpha}")
        print("=> Significant difference between sample means.")
    else:
        print(f"Fail to reject null hypothesis at α = {alpha}")
        print("=> No significant difference between sample means.")

# Simulate random data
np.random.seed(42)
pop_mean1, pop_mean2 = 50, 52
pop_std1, pop_std2 = 10, 10  # Assume known
sample_size1, sample_size2 = 40, 35

sample1 = np.random.normal(loc=pop_mean1, scale=pop_std1, size=sample_size1)
sample2 = np.random.normal(loc=pop_mean2, scale=pop_std2, size=sample_size2)

# Perform the two-sample Z-test
two_sample_z_test(sample1, sample2, pop_std1, pop_std2)
```

3.Implement a one-sample Z-test using Python to compare the sample mean with the population mean.

```python
import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample, population_mean, population_std, alpha=0.05):
    """
    Perform a one-sample Z-test to compare sample mean to a known population mean.

    Parameters:
        sample (array-like): The sample data
        population_mean (float): Known population mean
        population_std (float): Known population standard deviation
        alpha (float): Significance level (default = 0.05)

    Returns:
        z (float): Z-score
        p (float): P-value
    """
    n = len(sample)
    sample_mean = np.mean(sample)

    # Calculate Z statistic
    z = (sample_mean - population_mean) / (population_std / np.sqrt(n))

    # Two-tailed test p-value
    p = 2 * (1 - norm.cdf(abs(z)))

    print(f"Sample mean = {sample_mean:.4f}")
    print(f"Population mean = {population_mean}")
    print(f"Z-score = {z:.4f}")
    print(f"P-value = {p:.4f}")

    if p < alpha:
        print(f"Reject the null hypothesis at α = {alpha}")
        print("=> Sample mean is significantly different from the population mean.")
    else:
        print(f"Fail to reject the null hypothesis at α = {alpha}")
        print("=> No significant difference between sample and population mean.")

    return z, p

# Example usage
if __name__ == "__main__":
    np.random.seed(0)

    # Simulated sample data
    sample_data = np.random.normal(loc=52, scale=10, size=30)

    # Known population parameters
    population_mean = 50
    population_std = 10  # Known population standard deviation

    # Perform Z-test
    one_sample_z_test(sample_data, population_mean, population_std)
```

4.Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def two_tailed_z_test(sample, pop_mean, pop_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)
    z = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))
    p_value = 2 * (1 - norm.cdf(abs(z)))

    # Print test results
    print(f"Sample mean = {sample_mean:.4f}")
    print(f"Z-score = {z:.4f}")
    print(f"P-value = {p_value:.4f}")

    z_critical = norm.ppf(1 - alpha/2)
    print(f"Z-critical (±) = ±{z_critical:.2f}")

    if abs(z) > z_critical:
        print("Reject the null hypothesis (significant difference)")
    else:
        print("Fail to reject the null hypothesis (no significant difference)")

    # Visualization
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x, 0, 1)

    plt.figure(figsize=(10, 5))
    plt.plot(x, y, label='Standard Normal Distribution', color='blue')

    # Shade rejection regions
    plt.fill_between(x, y, where=(x <= -z_critical), color='red', alpha=0.4, label='Rejection Region (α/2)')
    plt.fill_between(x, y, where=(x >= z_critical), color='red', alpha=0.4)

    # Shade acceptance region
    plt.fill_between(x, y, where=(x > -z_critical) & (x < z_critical), color='green', alpha=0.2, label='Acceptance Region')

    # Plot Z-score
    plt.axvline(z, color='black', linestyle='--', label=f'Z-score = {z:.2f}')

    plt.title('Two-Tailed Z-Test: Decision Regions')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
np.random.seed(0)
sample_data = np.random.normal(loc=52, scale=10, size=30)
pop_mean = 50
pop_std = 10  # known
alpha = 0.05

two_tailed_z_test(sample_data, pop_mean, pop_std, alpha)
```

5.Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(pop_mean_H0, pop_mean_H1, pop_std, n, alpha=0.05):
    """
    Visualizes Type I and Type II errors in hypothesis testing.

    Parameters:
        pop_mean_H0 (float): Mean under the null hypothesis (H0)
        pop_mean_H1 (float): Mean under the alternative hypothesis (H1)
        pop_std (float): Population standard deviation (assumed known)
        n (int): Sample size
        alpha (float): Significance level
    """
    # Standard error
    se = pop_std / np.sqrt(n)

    # Critical value for one-tailed right-side test (can modify for two-tailed)
    z_critical = norm.ppf(1 - alpha)

    # Critical sample mean threshold
    critical_value = pop_mean_H0 + z_critical * se

    # Beta: P(Type II Error) = P(X < critical | H1 true)
    beta = norm.cdf((critical_value - pop_mean_H1) / se)

    # Plotting
    x = np.linspace(pop_mean_H0 - 4*se, pop_mean_H1 + 4*se, 1000)
    y_H0 = norm.pdf(x, pop_mean_H0, se)
    y_H1 = norm.pdf(x, pop_mean_H1, se)

    plt.figure(figsize=(10, 6))

    # H0 distribution
    plt.plot(x, y_H0, label='H₀: μ = {}'.format(pop_mean_H0), color='blue')
    plt.fill_between(x, y_H0, where=(x > critical_value), color='red', alpha=0.3, label='Type I Error (α)')

    # H1 distribution
    plt.plot(x, y_H1, label='H₁: μ = {}'.format(pop_mean_H1), color='green')
    plt.fill_between(x, y_H1, where=(x < critical_value), color='orange', alpha=0.3, label='Type II Error (β)')

    # Critical threshold
    plt.axvline(critical_value, color='black', linestyle='--', label='Critical Value')

    plt.title("Type I and Type II Error Visualization (One-tailed Z-test)")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

    # Display numerical values
    print(f"Critical value (sample mean threshold): {critical_value:.4f}")
    print(f"Z-critical: {z_critical:.4f}")
    print(f"Type II error (β): {beta:.4f}")
    print(f"Power of the test (1 - β): {1 - beta:.4f}")
```

```python
import numpy as np
from scipy.stats import ttest_ind

def independent_t_test(sample1, sample2, alpha=0.05, equal_var=True):
    """
    Perform an independent two-sample t-test and interpret the results.

    Parameters:
        sample1 (array-like): First sample data
        sample2 (array-like): Second sample data
        alpha (float): Significance level (default is 0.05)
        equal_var (bool): Assume equal population variance (default is True)
    """
    t_stat, p_value = ttest_ind(sample1, sample2, equal_var=equal_var)

    print("Independent T-Test Results:")
    print(f"Sample 1 mean = {np.mean(sample1):.4f}, n = {len(sample1)}")
    print(f"Sample 2 mean = {np.mean(sample2):.4f}, n = {len(sample2)}")
    print(f"T-statistic = {t_stat:.4f}")
    print(f"P-value = {p_value:.4f}")

    if p_value < alpha:
        print(f"Result: Reject the null hypothesis at α = {alpha}")
        print("=> There is a significant difference between the two sample means.")
    else:
        print(f"Result: Fail to reject the null hypothesis at α = {alpha}")
        print("=> No significant difference between the two sample means.")

# Example usage:
if __name__ == "__main__":
    np.random.seed(0)

    # Simulate two samples
    group1 = np.random.normal(loc=55, scale=10, size=30)
    group2 = np.random.normal(loc=50, scale=10, size=30)

    # Perform t-test
    independent_t_test(group1, group2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_rel

def paired_t_test(sample_before, sample_after, alpha=0.05):
    """
    Performs a paired sample T-test and visualizes the results.

    Parameters:
        sample_before (array-like): Measurements before (e.g., treatment)
        sample_after (array-like): Measurements after
        alpha (float): Significance level
    """
    # Calculate differences
    differences = np.array(sample_after) - np.array(sample_before)

    # Perform paired t-test
    t_stat, p_value = ttest_rel(sample_after, sample_before)

    # Print results
    print("Paired Sample T-Test Results:")
    print(f"Mean Before = {np.mean(sample_before):.4f}")
    print(f"Mean After  = {np.mean(sample_after):.4f}")
    print(f"T-statistic = {t_stat:.4f}")
    print(f"P-value     = {p_value:.4f}")

    if p_value < alpha:
        print("=> Reject the null hypothesis: Significant change detected.")
    else:
        print("=> Fail to reject the null hypothesis: No significant change.")

    # Visualization
    plt.figure(figsize=(10, 6))
    plt.plot([1]*len(sample_before), sample_before, 'o', label='Before')
    plt.plot([2]*len(sample_after), sample_after, 'o', label='After')

    # Connect paired samples
    for i in range(len(sample_before)):
        plt.plot([1, 2], [sample_before[i], sample_after[i]], 'k-', alpha=0.5)

    plt.xticks([1, 2], ['Before', 'After'])
    plt.ylabel('Measurement')
    plt.title('Paired Sample Comparison')
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

# Example usage
if __name__ == "__main__":
    np.random.seed(42)

    # Simulate paired data (e.g., pre and post treatment)
    before = np.random.normal(loc=70, scale=8, size=25)
    after = before + np.random.normal(loc=-3, scale=5, size=25)

    # Perform paired T-test and plot
    paired_t_test(before, after)
```

8.Simulate data and perform both Z-test and T-test, then compare the results using Python.

```python
import numpy as np
from scipy.stats import norm, t
import matplotlib.pyplot as plt

def simulate_and_compare_tests(pop_mean=100, pop_std=15, sample_size=25, alpha=0.05):
    np.random.seed(0)

    # Simulate sample from population
    sample = np.random.normal(loc=pop_mean, scale=pop_std, size=sample_size)
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)

    # Z-test (population std known)
    z_stat = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))
    p_z = 2 * (1 - norm.cdf(abs(z_stat)))

    # T-test (population std unknown)
    t_stat = (sample_mean - pop_mean) / (sample_std / np.sqrt(sample_size))
    df = sample_size - 1
    p_t = 2 * (1 - t.cdf(abs(t_stat), df))

    # Display results
    print("=== Test Results ===")
    print(f"Sample mean = {sample_mean:.4f}, Sample std = {sample_std:.4f}")
    print(f"Z-test: Z = {z_stat:.4f}, p-value = {p_z:.4f}")
    print(f"T-test: t = {t_stat:.4f}, p-value = {p_t:.4f}")

    if p_z < alpha:
        print("Z-test: Reject H₀")
    else:
        print("Z-test: Fail to reject H₀")

    if p_t < alpha:
        print("T-test: Reject H₀")
    else:
        print("T-test: Fail to reject H₀")

    # Plot both distributions
    x = np.linspace(-4, 4, 500)
    plt.figure(figsize=(10, 5))
    plt.plot(x, norm.pdf(x), label='Z-distribution', color='blue')
    plt.plot(x, t.pdf(x, df), label=f'T-distribution (df={df})', color='green')

    plt.axvline(z_stat, color='blue', linestyle='--', label=f'Z = {z_stat:.2f}')
    plt.axvline(t_stat, color='green', linestyle='--', label=f't = {t_stat:.2f}')

    plt.title('Comparison of Z-test and T-test Statistics')
    plt.xlabel('Test Statistic')
    plt.ylabel('Density')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Run the comparison
simulate_and_compare_tests()
```

9.Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
import scipy.stats as stats
import numpy as np

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
    - data (list or array): The sample data.
    - confidence (float): The confidence level (default is 0.95 for 95%).

    Returns:
    - A tuple (lower_bound, upper_bound) representing the confidence interval.
    """
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # Standard error of the mean
    margin = std_err * stats.t.ppf((1 + confidence) / 2.0, df=n-1)

    return mean - margin, mean + margin
```

```python
sample_data = [12, 15, 14, 10, 13, 15, 11, 14, 12, 13]
ci = confidence_interval(sample_data, confidence=0.95)
print("95% Confidence Interval:", ci)
```

10.Write a Python program to calculate the margin of error for a given confidence level using sample data.

```python
import scipy.stats as stats
import numpy as np

def calculate_margin_of_error(data, confidence=0.95):
    """
    Calculate the margin of error for the sample mean at a given confidence level.

    Parameters:
    - data (list or array): Sample data
    - confidence (float): Confidence level (e.g., 0.95 for 95%)

    Returns:
    - margin_of_error (float): The margin of error
    """
    data = np.array(data)
    n = len(data)
    std_err = stats.sem(data)  # Standard error of the mean
    t_score = stats.t.ppf((1 + confidence) / 2.0, df=n - 1)
    margin_of_error = t_score * std_err
    return margin_of_error
```

```python
sample_data = [50, 52, 49, 48, 51, 53, 47, 50, 52, 49]
confidence_level = 0.95

moe = calculate_margin_of_error(sample_data, confidence_level)
print(f"Margin of Error at {confidence_level*100:.0f}% confidence level: {moe:.2f}")
```

11.Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

```python
def bayes_theorem(prior_H, likelihood_D_given_H, likelihood_D_given_not_H):
    """
    Calculate the posterior probability using Bayes' Theorem.

    Parameters:
    - prior_H: P(H), prior probability of the hypothesis
    - likelihood_D_given_H: P(D|H), likelihood of data given the hypothesis
    - likelihood_D_given_not_H: P(D|¬H), likelihood of data given not the hypothesis

    Returns:
    - posterior: P(H|D), updated probability of hypothesis given data
    """
    prior_not_H = 1 - prior_H
    evidence = (likelihood_D_given_H * prior_H) + (likelihood_D_given_not_H * prior_not_H)
    posterior = (likelihood_D_given_H * prior_H) / evidence
    return posterior
```

```python
# Parameters for the test
prior = 0.01                # 1% of people have the disease
sensitivity = 0.99          # Test detects disease correctly
false_positive = 0.05       # Test falsely detects disease in healthy people

posterior = bayes_theorem(prior, sensitivity, false_positive)
print(f"Probability of disease given a positive test: {posterior:.4f}")
```

12.Perform a Chi-square test for independence between two categorical variables in Python.

```python
import pandas as pd
import scipy.stats as stats

# Step 1: Create a contingency table
# Rows: Gender, Columns: Product Preference
data = {
    'Product A': [30, 20],
    'Product B': [20, 30]
}
contingency_table = pd.DataFrame(data, index=['Male', 'Female'])

# Step 2: Perform Chi-square test
chi2, p, dof, expected = stats.chi2_contingency(contingency_table)

# Step 3: Print the results
print("Chi-square Statistic:", chi2)
print("Degrees of Freedom:", dof)
print("P-value:", p)
print("\nExpected Frequencies:\n", expected)

# Step 4: Interpret the result
alpha = 0.05
if p < alpha:
    print("\n🔔 Reject the null hypothesis: There is a significant association between Gender and Preference.")
else:
    print("\n✅ Fail to reject the null hypothesis: No significant association between Gender and Preference.")
```

13.Write a Python program to calculate the expected frequencies for a Chi-square test based on observed
data.

```python
import numpy as np

def calculate_expected_frequencies(observed):
    """
    Calculate expected frequencies for a Chi-square test.

    Parameters:
    - observed (2D list or numpy array): Observed frequency table

    Returns:
    - expected (numpy array): Expected frequency table
    """
    observed = np.array(observed)
    row_sums = observed.sum(axis=1).reshape(-1, 1)  # Sum of each row (as column vector)
    col_sums = observed.sum(axis=0).reshape(1, -1)  # Sum of each column (as row vector)
    total = observed.sum()

    expected = (row_sums @ col_sums) / total  # Matrix multiplication to get expected frequencies
    return expected

# Example usage:
observed_data = [
    [30, 20],
    [20, 30]
]

expected_freq = calculate_expected_frequencies(observed_data)
print("Expected Frequencies:")
print(expected_freq)
```

14.Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

```python
import scipy.stats as stats

def goodness_of_fit_test(observed, expected_probs):
    """
    Perform Chi-square goodness-of-fit test.

    Parameters:
    - observed (list or array): Observed counts for each category
    - expected_probs (list or array): Expected probabilities for each category (should sum to 1)

    Returns:
    - chi2_stat: Chi-square statistic
    - p_value: p-value of the test
    """
    observed = list(observed)
    expected_probs = list(expected_probs)

    total = sum(observed)
    expected = [total * p for p in expected_probs]

    chi2_stat, p_value = stats.chisquare(f_obs=observed, f_exp=expected)
    return chi2_stat, p_value

# Example usage:
observed_counts = [50, 30, 20]  # Observed frequencies in 3 categories
expected_distribution = [0.4, 0.4, 0.2]  # Expected probabilities

chi2, p = goodness_of_fit_test(observed_counts, expected_distribution)
print(f"Chi-square Statistic: {chi2:.3f}")
print(f"P-value: {p:.4f}")

alpha = 0.05
if p < alpha:
    print("Reject the null hypothesis: Observed data does NOT fit the expected distribution.")
else:
    print("Fail to reject the null hypothesis: Observed data fits the expected distribution.")
```
